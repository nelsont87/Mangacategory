```python
import pandas as pd
```

```python
top_fifty = pd.read_csv("01_page_top50mangaMAL.csv")
top_fifty
```

```python
# break apart the list within the column then add it to new columns in the dataframe
#trying out explode
genres_df = top_fifty[['Genres', 'English Title']]
genres_df
```

```python
#testing to see the format of the column
genres_df.iloc[0,0]
```

```python
from ast import literal_eval

genres_df['Genres'] = genres_df['Genres'].apply(literal_eval)
```

```python
exploded_df = genres_df.explode('Genres')
exploded_df
```

```python
#getting the counts of the different genres
top = exploded_df["Genres"].value_counts()
top
```

```python
top.plot(kind = "bar")
```

```python
next_fifty = pd.read_csv("02_page_top50-100mangaMAL.csv")
next_fifty
```

```python
genres_next_df = next_fifty[['Genres', 'English Title']]
genres_next_df
```

```python
#converting the column to an actual list
genres_next_df['Genres'] = genres_next_df['Genres'].apply(literal_eval)
```

```python
#exploding the column to get the different genres
exploded_next_df = genres_next_df.explode('Genres')
exploded_next_df
```

```python
#getting the counts of the different genres
nxt = exploded_next_df["Genres"].value_counts()
nxt
```

```python
nxt.plot(kind = "bar")
```
